```python


import scrapy
from scrapy.crawler import CrawlerProcess

class GoogleScholarSpider(scrapy.Spider):
    name = 'google_scholar'
    start_urls = ['https://scholar.google.com/scholar?q=Blockchain+and+privacy+protection']

    custom_settings = {
        'FEED_FORMAT': 'csv',
        'FEED_URI': 'output.csv',
        'ROBOTSTXT_OBEY': True,
        'USER_AGENT': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.159 Safari/537.36'
    }

    def parse(self, response):
        # 提取论文标题和链接
        for article in response.css('div.gs_r.gs_or.gs_scl'):
            yield {
                'title': article.css('h3.gs_rt a::text').extract_first(),
                'link': article.css('h3.gs_rt a::attr(href)').extract_first(),
            }

        # 如果还没有爬取40篇，继续爬取下一页
        next_page = response.css('div.gs_n a::attr(href)').extract()[-1]
        if next_page is not None:
            yield response.follow(next_page, self.parse)

# 运行爬虫
process = CrawlerProcess()
process.crawl(GoogleScholarSpider)
process.start()
```

```python
import pandas as pd
from scholarly import scholarly

def search_papers_on_google_scholar(df):
    for index, row in df.iterrows():
        title = row['题目']
        print(f"Searching for: {title}")
        
        # Use scholarly to search for the paper on Google Scholar
        search_query = scholarly.search_pubs(title)
        try:
            paper = next(search_query)
            print(f"Found: {paper.bib['title']}")
            print(f"Link: {paper.bib['url']}\n")
        except StopIteration:
            print(f"No results found for: {title}\n")
        
# Load the excel file into a dataframe
df = pd.read_excel("Blockchain data obfuscation.xlsx")

# Use the function to search papers on Google Scholar
search_papers_on_google_scholar(df)

```
